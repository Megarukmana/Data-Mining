```python
weather['MinTemp']=weather['MinTemp'].fillna((weather['MinTemp'].median()))
weather['MaxTemp']=weather['MaxTemp'].fillna((weather['MaxTemp'].median()))
weather['Rainfall']=weather['Rainfall'].fillna((weather['Rainfall'].median()))
weather['Evaporation']=weather['Evaporation'].fillna((weather['Evaporation'].median()))
weather['Sunshine']=weather['Sunshine'].fillna((weather['Sunshine'].median()))
weather['WindGustSpeed']=weather['WindGustSpeed'].fillna((weather['WindGustSpeed'].median()))
weather['WindSpeed9am']=weather['WindSpeed9am'].fillna((weather['WindSpeed9am'].median()))
weather['WindSpeed3pm']=weather['WindSpeed3pm'].fillna((weather['WindSpeed3pm'].median()))
weather['Humidity9am']=weather['Humidity9am'].fillna((weather['Humidity9am'].median()))
weather['Humidity3pm']=weather['Humidity3pm'].fillna((weather['Humidity3pm'].median()))
weather['Pressure9am']=weather['Pressure9am'].fillna((weather['Pressure9am'].median()))
weather['Pressure3pm']=weather['Pressure3pm'].fillna((weather['Pressure3pm'].median()))
weather['Cloud9am']=weather['Cloud9am'].fillna((weather['Cloud9am'].median()))
weather['Cloud3pm']=weather['Cloud3pm'].fillna((weather['Cloud3pm'].median()))
weather['Temp9am']=weather['Temp9am'].fillna((weather['Temp9am'].median()))
weather['Temp3pm']=weather['Temp3pm'].fillna((weather['Temp3pm'].median()))
weather['RISK_MM']=weather['RISK_MM'].fillna((weather['RISK_MM'].median()))
np.sum(weather.isnull())
```

```python
weather_tanpa_missing_value= weather
print( weather_tanpa_missing_value)
```

## 2. Missing Value dan Outlier

```python
from scipy import stats
weather1=weather_tanpa_missing_value.drop(['WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow'], 1)
z = np.abs(stats.zscore(weather1._get_numeric_data()))
print(z)
```

```python
weather_tanpa_missing_value_dan_outlier = weather_tanpa_missing_value[(z < 3).all(axis=1)]
print(weather_tanpa_missing_value_dan_outlier.shape)
```

```python
weather_tanpa_missing_value_dan_outlier.head()
```

## Transformasi Data

```python
weather2= weather_tanpa_missing_value_dan_outlier.drop(['Date', 'Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow'], 1)
from sklearn.preprocessing import normalize
norm = normalize(weather2, norm ='l2')
norm = pd.DataFrame(norm)
norm
```

## 4. feature Selection (Mencari Skor)

```python
from sklearn.feature_selection import SelectKBest, chi2, f_regression # metode feature selection, chi2 untuk klasifikasi
weather3= weather_tanpa_missing_value_dan_outlier.drop(['Date', 'Location'], 1)
x = weather3.drop(columns=['RainTomorrow'])
y = weather_tanpa_missing_value_dan_outlier[['RainTomorrow']]
selector = SelectKBest (f_regression, k='all').fit(x,y) #memunculkan 1 variabel prediktor terpenting
X_new = selector.transform(x)
scores = selector.scores_
scores
```
