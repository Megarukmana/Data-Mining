# Mendapatkan Model Regresi Terbaik dengan Preprocessing Data
## Data yang digunakan merupakan data indeks kebahagiaan Indonesia tahun 2014 yang didapatkan dari BPS. Indeks kebahagiaan merupakan indeks komposit tertimbang yang mencakup 10 indikator kepuasan penduduk dan juga sebagai ukuran tingkat kebahagiaan penduduk. 

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('D:\statistika\SEMESTER 6\DATMIN\Transformasi Data\Indeks Kebahagiaan Menurut Provinsi Tahun 2014.csv', sep=';')
data
```

```python
data.describe()
```

## a. Missing Value

```python
np.sum(data.isnull())
```

```python
data['Indeks Kebahagiaan']=data['Indeks Kebahagiaan'].fillna((data['Indeks Kebahagiaan'].mean()))
data['Kesehatan']=data['Kesehatan'].fillna((data['Kesehatan'].mean()))
data['Pendidikan']=data['Pendidikan'].fillna((data['Pendidikan'].mean()))
data['Keharmonisan Keluarga']=data['Keharmonisan Keluarga'].fillna((data['Keharmonisan Keluarga'].mean()))
data['Ketersediaan Waktu Luang']=data['Ketersediaan Waktu Luang'].fillna((data['Ketersediaan Waktu Luang'].mean()))
data['Hubungan Sosial']=data['Hubungan Sosial'].fillna((data['Hubungan Sosial'].mean()))
data['Keadaan Lingkungan']=data['Keadaan Lingkungan'].fillna((data['Keadaan Lingkungan'].mean()))
data['Kondisi Keamanan']=data['Kondisi Keamanan'].fillna((data['Kondisi Keamanan'].mean()))
data['Pekerjaan']=data['Pekerjaan'].fillna((data['Pekerjaan'].mean()))
data['Pendapatan Rumah Tangga']=data['Pendapatan Rumah Tangga'].fillna((data['Pendapatan Rumah Tangga'].mean()))
data['Kondisi Rumah dan Aset']=data['Kondisi Rumah dan Aset'].fillna((data['Kondisi Rumah dan Aset'].mean()))
np.sum(data.isnull())
```

```python
data_tanpa_missing_value=data
print(data_tanpa_missing_value)
```

```python
data_tanpa_missing_value.to_excel('a. Data Indeks Kebahagiaan tanpa Missing Value.xlsx')
```

## b. Missing Value dan Outlier

```python
from scipy import stats
z = np.abs(stats.zscore(data_tanpa_missing_value._get_numeric_data()))
print(z)
data_tanpa_missing_value_dan_outlier =data_tanpa_missing_value[(z<3).all(axis=1)]
print(data_tanpa_missing_value_dan_outlier.shape)
```

```python
data_tanpa_missing_value_dan_outlier.head()
```

```python
data_tanpa_missing_value_dan_outlier.tail()
```

```python
data_tanpa_missing_value_dan_outlier.to_excel('b. Data Indeks Kebahagiaan tanpa Missing Value dan Outlier.xlsx')
```

## c. Missing Value dan Transformasi Data

```python
x = data_tanpa_missing_value.iloc[0:35,1:12]
x.head()
```

```python
from sklearn.preprocessing import normalize
norm = normalize(x, norm ='l2')
norm = pd.DataFrame(norm)
norm.head()
```

```python
norm.tail()
```

```python
transformasi_sigmoidal = (1-2.718281828**(-norm))/(1+2.718281828**(-norm))
transformasi_sigmoidal = pd.DataFrame(transformasi_sigmoidal)
transformasi_sigmoidal.head()
```

```python
transformasi_sigmoidal.tail()
```

```python
transformasi_sigmoidal.to_excel('c. Data Indeks Kebahagiaan tanpa Missing Value dan Transformasi.xlsx')
```

## d. Missing Value, Outlier, dan Transformasi Data

```python
y = data_tanpa_missing_value_dan_outlier.iloc[0:35,1:12]
y.head()
```

```python
from sklearn.preprocessing import normalize
ynorm = normalize(y, norm ='l2')
ynorm = pd.DataFrame(ynorm)
ynorm.head()
```

```python
transformasi_sigmoidal_y = (1-2.718281828**(-ynorm))/(1+2.718281828**(-ynorm))
transformasi_sigmoidal_y = pd.DataFrame(transformasi_sigmoidal_y)
transformasi_sigmoidal_y.head()
```

```python
transformasi_sigmoidal_y.to_excel('d. Data Indeks Kebahagiaan tanpa Missing Value, outlier dan Transformasi.xlsx')
```
