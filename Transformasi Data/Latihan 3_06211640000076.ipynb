# Transformasi Data

```python
import pandas as pd
import numpy as np
```

```python
#import datase
pcademo=pd.read_csv("pcademo.csv", sep=',')
x=pcademo.iloc[0:51,1:20]
x.tail()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm 12, other are L1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
# rentangnya antara 0 sampai 1
```

```python
#minmax
#dimana rentang nilai bisa kita tentuin sendiri
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-2,1))# nilainya antara -2 sampai 1
scaler.fit(x)
x_scaled=scaler.transform(x) #x_scaled : nama
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
#3sigmoid
xsigmoidal=(1-2.718281828**(-xnorm))/(1+2.718281828**(-xnorm)) #nilai eksponensial =2.718281828
xsigmoidal=pd.DataFrame(xsigmoidal) 
xsigmoidal.tail()
```

```python
# decimal scalling
xdecimal=x/(10**3)# 10** i, i ditentukan peneliti
xdecimal=pd.DataFrame(xdecimal)
xdecimal.tail()
```

```python
#softmax
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
#
```

```python
# feature selection (mencari skor)
from sklearn.feature_selection import SelectKBest, chi2, f_regression
Xx = x.iloc[0:51,2:20]
y = x.iloc[0:51,1]
selector = SelectKBest (f_regression, k="all").fit(Xx,y)
X_new = selector.transform(Xx)
scores = selector.scores_
scores
#semakin tinggi skore, maka variabel semakin baik untuk model
```
